```python
import numpy as np
import matplotlib.pyplot as plt
from math import *



## B∆∞·ªõc 1: ƒê·ªçc file d·ªØ li·ªáu

def get_input(file_can_doc):
    X_data = np.genfromtxt(file_can_doc, delimiter = ',', skip_header = 1, dtype= str);
    ## N·∫øu m√† ta kh√¥ng khai b√°o trong th∆∞ vi·ªán genfromtxt th√¨ c√°c d·ªØ li·ªáu ƒë∆∞·ª£c ƒë·ªçc v√†o d∆∞·ªõi d·∫°ng s·ªë
    ## M√† label th√¨ ·ªü d·∫°ng str n√™n ta c·∫ßn √©p khai dtype = str trong h√†m genfromtxt ƒë·ªÉ h√†m ƒë·ªçc ƒë∆∞·ª£c chu·ªói
    ## H∆°n n·ªØa, n·∫øu l√†m th√¨ s·∫Ω ·∫£nh h∆∞·ªüng ƒë·∫øn c√°c d·ªØ li·ªáu s·ªë trong c√°c c·ªôt kh√°c, v√¨ ·ªü b∆∞·ªõc t√≠nh distance gi·ªØa c√°c ƒëi·ªÉm m·∫´u (h√†ng x√≥m) v·ªõi item
    ## Th√¨ ta c·∫ßn √©p ki·ªÉu sang float ƒë·ªÉ t√≠nh to√°n.

    trainingSet = X_data[:149, 1:]
    testingSet = X_data[149:, 1:-1]  ## L∆∞u √Ω: B·ªè ƒëi lable - v√¨ label n√†y c·∫ßn ƒë∆∞·ª£c d·ª± ƒëo√°n

    return trainingSet, testingSet

## B∆∞·ªõc 2: Vi·∫øt h√†m t√≠nh kho·∫£ng c√°ch gi·ªØa hai ƒëi·ªÉm
def computeDistance(dataPoint1, dataPoint2):
    result = 0;
    for i in range(4):
        result += (float(dataPoint1[i]) - float(dataPoint2[i])) ** 2
    result = sqrt(result)
    return result

## B∆∞·ªõc 3: Vi·∫øt h√†m ch·ªçn ra K h√†ng x√≥m l√°ng gi·∫øng g·∫ßn nh·∫•t c·ªßa iteam c·∫ßn d·ª± ƒëo√°n nh√£n

def computeKnearestNeighbor(trainingSet, item, k):
    distances = []
    for dataPoint in trainingSet:
        distances.append(
            { "label": dataPoint[-1],
               "value": computeDistance(dataPoint, item)
            }
        )

    distances.sort(key = lambda x: x["value"])
    labels = [item["label"] for item in distances]

    return labels[:k]

## B∆∞·ªõc 4: Vi·∫øt h√†m vote ƒë·ªÉ d·ª± ƒëo√°n nh√£n c·ªßa item th√¥ng qua c·ªßa K h√†ng x√≥m g·∫ßn nh·∫•t

def voteTheDistances(array):
    labels = set(array)
    result = ""
    maxOccur = 0
    for label in labels:
        num = array.count(label)
        if (num > maxOccur):
            maxOccur = num
            result = label

    return result


###-------------------------CH∆Ø∆†NG TR√åNH CH√çNH-----------------------------###
if __name__ == "__main__":
    k = 5;
    trainingSet, testingSet = get_input("Iris(1).csv")
    for item in testingSet:
        labels = computeKnearestNeighbor(trainingSet, item, k)
        label_prediction =  voteTheDistances(labels)
        print(f"Nh√£n d·ªØ li·ªáu c·∫ßn d·ª± ƒëo√°n cho {item} l√†: {label_prediction}")
    
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

## B∆∞·ªõc 1: Ch·ªçn h√†m ƒë·ªçc v√† x·ª≠ l√Ω d·ªØ li·ªáu ƒë·∫ßu v√†o b·∫±ng pandas
def normalized_data(X_data):
    n = len(X_data);
    m = len(X_data[0])
    

    # T√≠nh trung b√¨nh m·ªói c·ªôt
    mean_data = [0] * m
    for j in range(m):
        for i in range(n):
            mean_data[j] += X_data[i][j]
        mean_data[j] /= n

    # T√≠nh ƒë·ªô l·ªánh chu·∫©n theo m·ªói c·ªôt
    std_data = [0] * m
    for j in range(m):
        for i in range(n):
            std_data[j] += (X_data[i][j] - mean_data[j]) ** 2
        std_data[j] = (std_data[j] / (n - 1)) ** 0.5  ## sqrt(S / (n - 1__
    
    ## Chu·∫©n h√≥a Z-score cho t·ª´ng ph·∫ßn t·ª≠

    X_normalized = []
    for i in range(n):
        row = []
        for j in range(m):
            if (std[j] != 0):
                row.append((X_data[i][j] - mean_data[j]) / std_data[j])
            else:
                row.append(0) # Tr√°nh chia cho 0

        X_normalized.append(row)


    return X_normalized
    
def get_input_by_pandas(file_can_doc):
    ## ƒê·ªçc file csv v√† ƒë·∫∑t l·∫°i ti√™u ƒë·ªÅ
    names = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']
    df = pd.read_csv(file_can_doc, names = names, header = 0)
    X_data = df.iloc[:, :-1].values
    y_data = df.iloc[:, -1].values

    ## Ph√¢n th√†nh c√°c t·∫≠p train, test ƒë·ªëi v·ªõi features v√† label
    X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.2, random_state = 42)
    
    '''
    X_train, X_test c√≥ ki·ªÉu d·ªØ li·ªáu l√† NumPy array 2D
    y_train, y_test c√≥ ki·ªÉu d·ªØ li·ªáu l√† NumPy array 1D
    '''

    ## Ti·∫øn h√†nh chu·∫©n h√≥a d·ªØ li·ªáu b·∫±ng h√†m t·ª± ƒë·ªãnh nghƒ©a ph√≠a tr√™n
    # X_train_normalize = normalized_data(X_train)
    # X_test_normalize = normalized_data(X_test)

    ## Ti·∫øn h√†nh chu·∫©n h√≥a d·ªØ li·ªáu b·∫±ng th∆∞ vi·ªán c·ªßa sklearn
    scaler = StandardScaler()
    scaler.fit(X_train)

    X_train_normalize = scaler.transform(X_train)
    X_test_normalize = scaler.transform(X_test)


    return X_train_normalize, X_test_normalize, y_train, y_test

from sklearn.neighbors import KNeighborsClassifier

def KNN(n_neighbors, X_train, y_train):
    classifier = KNeighborsClassifier(n_neighbors = n_neighbors, weights = 'uniform')
    classifier.fit(X_train, y_train)

    return classifier


    

if __name__ == "__main__":
    n_neighbors = 5;
    X_train_normalize, X_test_normalize, y_train, y_test = get_input_by_pandas("iris(1).csv")
    classifier = KNN(n_neighbors, X_train_normalize, y_train)

    y_pred = classifier.predict(X_test_normalize)
    print(y_pred != y_test)


    from sklearn.metrics import accuracy_score, classification_report
    print(round(accuracy_score(y_test, y_pred), 2))
    print(classification_report(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors, datasets
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report
```

```python
#---Demo---#
iris = np.genfromtxt("Iris(1).csv", delimiter = ',', skip_header = 1, dtype = str)
iris_X = iris[:, :-1]
iris_X = iris_X.astype(float)
iris_y = iris[:, -1]
'''
np.unique(iris) --> Tr·∫£ v·ªÅ s·ªë l·ªõp ph√¢n lo·∫°i (s·ªë nh√£n kh√°c nhau) c·ªßa iris
len(iris) --> Tr·∫£ v·ªÅ s·ªë l∆∞·ª£ng chi·ªÅu d√†i c·ªßa m·∫´u d·ªØ li·ªáu (s·ªë d√≤ng d·ªØ li·ªáu)
'''
print('Number of classes: %d' %len(np.unique(iris_y)))
print('Number of data points: %d' %len(iris_y))

```

```python
from sklearn.model_selection import train_test_split
'''
√Ä! C√¢u h·ªèi c·ªßa b·∫°n l√†: C√≥ c·∫ßn ph·∫£i g√°n ƒë√∫ng th·ª© t·ª± X_train, X_test, y_train, y_test khi d√πng train_test_split kh√¥ng?

üëâ C√¢u tr·∫£ l·ªùi: C√≥, b·∫°n ph·∫£i g√°n ƒë√∫ng th·ª© t·ª±.
'''
'''
test_size = a t·ª©c l√† d·ªØ li·ªáu tr√™n t·∫≠p test c√≥ a d√≤ng v√† s·ªë c·ªôt b·∫±ng s·ªë c·ªôt c·ªßa t·∫≠p train g·ªëc
'''
X_train, X_test, y_train, y_test = train_test_split(
    iris_X, iris_y, test_size = 50, random_state = 42)

print("Training size: %d" %len(y_train))
print("Test size    : %d" %len(y_test))
```

```python
clf = neighbors.KNeighborsClassifier(n_neighbors = 8, p = 2) 
## N·∫øu kh√¥ng khai b√°o weights = g√¨ trong h√†m neigbors.KNeighborsClassifier th√¨
## th∆∞ vi·ªán t·ª± hi·ªÉu l√† ch√∫ng ta ƒëang ƒë·ªÉ gi√° tr·ªã m·∫∑c ƒë·ªãnh weights = uniform
## Xem c√°c tr·ªçng s·ªë c·ªßa K h√†ng x√≥m l√°ng gi·ªÅng g·∫ßn item nh·∫•t l√† nh∆∞ nhau, v√† ch·ªçn ng·∫´u nhi√™n n·∫øu K l√† s·ªë ch·∫µn)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("K·∫øt qu·∫£ d·ª± ƒëo√°n cho 20 test data points:")
print("Predicted labels: ", y_pred[20:40])
print("Ground truth    : ", y_test[20:40])

print ("Accuracy of 10NN with major voting: %.2f %%" %(100*accuracy_score(y_test, y_pred)))
print(classification_report(y_test, y_pred))
```

# Hi·ªÉu v·ªÅ `weights='distance'` trong KNeighborsClassifier (scikit-learn)

Khi b·∫°n d√πng `weights='distance'` trong `KNeighborsClassifier`, √Ω nghƒ©a l√† **m·ªói l√°ng gi·ªÅng (neighbor) ƒë∆∞·ª£c g√°n tr·ªçng s·ªë t·ªâ l·ªá ngh·ªãch v·ªõi kho·∫£ng c√°ch ƒë·∫øn ƒëi·ªÉm c·∫ßn d·ª± ƒëo√°n**.

---

## ‚úÖ C·ª• th·ªÉ:

Trong KNN, ƒë·ªÉ d·ª± ƒëo√°n nh√£n cho m·ªôt ƒëi·ªÉm m·ªõi $ x $, ta t√¨m **k ƒëi·ªÉm g·∫ßn nh·∫•t** trong t·∫≠p hu·∫•n luy·ªán. 

- M·∫∑c ƒë·ªãnh (`weights='uniform'`): m·ªói ƒëi·ªÉm g√≥p **1 phi·∫øu b·∫±ng nhau**.
- V·ªõi `weights='distance'`: ƒëi·ªÉm g·∫ßn h∆°n s·∫Ω c√≥ **·∫£nh h∆∞·ªüng l·ªõn h∆°n**.

---

## üìê C√¥ng th·ª©c:

Gi·∫£ s·ª≠:
- $ x $: ƒëi·ªÉm c·∫ßn d·ª± ƒëo√°n  
- $ x_i $: ƒëi·ªÉm l√°ng gi·ªÅng g·∫ßn nh·∫•t th·ª© $ i $  
- $ d_i = \| x - x_i \| $: kho·∫£ng c√°ch t·ª´ $ x $ ƒë·∫øn $ x_i $  
- $ y_i $: nh√£n c·ªßa $ x_i $

**Tr·ªçng s·ªë ƒë∆∞·ª£c t√≠nh:**

$$
w_i = \frac{1}{d_i}
$$

T·ªïng tr·ªçng s·ªë cho t·ª´ng nh√£n ƒë∆∞·ª£c t√≠nh, sau ƒë√≥ **l·∫•y nh√£n c√≥ t·ªïng tr·ªçng s·ªë cao nh·∫•t** l√†m k·∫øt qu·∫£ d·ª± ƒëo√°n.

---

## üß† V√≠ d·ª• tr·ª±c quan:

Gi·∫£ s·ª≠ c√≥ 3 l√°ng gi·ªÅng g·∫ßn nh·∫•t nh∆∞ sau:

| L√°ng gi·ªÅng | Nh√£n | Kho·∫£ng c√°ch | Tr·ªçng s·ªë |
|------------|------|-------------|----------|
| A          | 0    | 0.1         | 10.0     |
| B          | 1    | 0.5         | 2.0      |
| C          | 1    | 0.2         | 5.0      |

- T·ªïng tr·ªçng s·ªë cho nh√£n 0: **10.0**  
- T·ªïng tr·ªçng s·ªë cho nh√£n 1: **2.0 + 5.0 = 7.0**

‚û°Ô∏è K·∫øt qu·∫£ d·ª± ƒëo√°n l√† **nh√£n 0**, v√¨ n√≥ c√≥ t·ªïng tr·ªçng s·ªë cao h∆°n.

---

## ‚úÖ Khi n√†o n√™n d√πng `weights='distance'`?

- Khi b·∫°n mu·ªën **∆∞u ti√™n l√°ng gi·ªÅng g·∫ßn h∆°n**.
- Gi√∫p gi·∫£m ·∫£nh h∆∞·ªüng c·ªßa **outlier ·ªü xa**.
- D√πng t·ªët v·ªõi **d·ªØ li·ªáu c√≥ nhi·ªÖu nh·∫π**.


```python
## KNN v·ªõi tr·ªçng s·ªë t·ª± ƒë·ªãnh nghƒ©a
def myweight(distances):
    sigma_2 = 0.5
    weights_i = np.exp(-distances ** 2/ sigma_2)
    return weights_i
    
clf = neighbors.KNeighborsClassifier(n_neighbors = 8, p = 2, weights = myweight)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("Accuracy of 10NN with major voting: %.2f %%" %(100 * accuracy_score(y_test, y_pred)))
print(classification_report(y_test, y_pred))
    
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline

```

```python
df = pd.read_csv("cruise_ship_info.csv")
df.head()

# ch·ªçn ƒë·∫∑c tr∆∞ng
cols_selected = ['Tonnage', 'passengers', 'length', 'cabins','crew']
X = df[cols_selected].iloc[:,0:4].values
y = df[cols_selected]['crew']

# Chu·∫©n h√≥a cho y
sc_y = StandardScaler()

## Pipeline
pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('slr', LinearRegression())])
pipe_knn = Pipeline([('scl', StandardScaler()),
                   ('knn', KNeighborsRegressor())])

## T·∫°o m·ªôt list ƒë·ªÉ l∆∞u ƒëi·ªÉm
train_score_lr = []
train_score_knn =  []

## S·ªë l·∫ßn l·∫∑p
n = 15

# ƒê√°nh gi√° m√¥ h√¨nh
for i in range(n):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,  random_state = i)
    
    y_train_std = sc_y.fit_transform(y_train.to_numpy().reshape(-1, 1)).flatten()

    
    train_score_lr = np.append(train_score_lr, np.mean(cross_val_score(pipe_lr, 
                                                                       X_train, y_train_std, 
                                                                        scoring = 'r2', cv = 10)))
    train_score_knn = np.append(train_score_knn, np.mean(cross_val_score(pipe_knn, 
                                                                         X_train, y_train_std,
                                                                        scoring = 'r2', cv = 10)))

# Chuy·ªÉn th√†nh array
train_score_lr = np.array(train_score_lr)
train_score_knn = np.array(train_score_knn)

# T√≠nh trung b√¨nh v√† ƒë·ªô l·ªách chu·∫©n
train_mean_lr = np.mean(train_score_lr)
train_std_lr = np.std(train_score_lr)
train_mean_knn = np.mean(train_score_knn)
train_std_knn = np.std(train_score_knn)
print('R2 train for lr: %.3f +/- %.3f' %
                             (train_mean_lr,train_std_lr))
print('R2 train for knn_lr: %.3f +/- %.3f' %
                             (train_mean_knn,train_std_knn))
```

```python
## Plot ƒë·ªì th·ªã hai m√¥ h√¨nh ƒë·ªÉ tr·ª±c quan h√≥a
import matplotlib.pyplot as plt

plt.figure(figsize=(15,11))
plt.plot(range(n),train_score_lr,color='blue', linestyle='dashed',
         marker='o',markerfacecolor='blue', markersize=10,
         label='linear regression')
plt.fill_between(range(n),
                 train_score_lr + train_std_lr,
                 train_score_lr - train_std_lr,
                 alpha=0.15, color='blue')
plt.plot(range(n),train_score_knn,color='green', linestyle='dashed',
         marker='s',markerfacecolor='green', markersize=10,
         label = 'Kneighbors regression')
plt.fill_between(range(n),
                 train_score_knn + train_std_knn,
                 train_score_knn - train_std_knn,
                 alpha=0.15, color='green')
plt.grid()
plt.ylim(0.7,1)
plt.title('Mean cross-validation R2 score vs. random state parameter', size = 14)
plt.xlabel('Random state parameter', size = 14)
plt.ylabel('Mean cross-validation R2 score', size = 14)
plt.legend()
plt.show()
```

```python
### T√¨nh hu·ªëng: trong nhi·ªÅu b√†i to√°n vi·ªác k l√† s·ªë ch·∫µn m√† ta kh√¥ng y√™u c·∫ßu
### t√≠nh tr·ªçng s·ªë cho t·ª´ng lables c·ªßa K h√†ng x√≥m g·∫ßn nh·∫•t th√¨ s·∫Ω d·∫´n t·ªõi vi·ªác 
### M√¥ h√¨nh ch·ªçn ng·∫´u nhi√™n 1 trong nhi·ªÅu lables c√≥ t·ªâ l·ªá b·∫±ng nhau --> D·∫´n ƒë·∫øn hi·ªáu su·∫•t m√¥ h√¨nh th·∫•p
### Sau ƒë√¢y l√† v√≠ d·ª• v·ªÅ ƒëi·ªÅu tr√™n


import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler


## B∆∞·ªõc 1: ƒê·ªçc d·ªØ li·ªáu
penguins = np.genfromtxt("penguins.csv", delimiter = ',', skip_header = 1, dtype = str)
X_data = penguins[:, 2:6]
X_data = X_data.astype(float)
y_data = penguins[:, 0]

## B∆∞·ªõc 2: X·ª≠ l√Ω d·ªØ li·ªáu
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data,
                                                    test_size = 0.2,
                                                    random_state = 42)
# B∆∞·ªõc 3: Chu·∫©n h√≥a d·ªØ li·ªáu
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)

# B∆∞·ªõc 4: Build 2 model ƒë·ªÉ so s√°nh
n_neighbors = list(range(1, 11))

scores_uniform = []
scores_distance = []

for k in n_neighbors:
    # KNN v·ªõi tr·ªçng s·ªë ƒë·ªÅu    
    knn = KNeighborsClassifier(n_neighbors = k, p = 2, weights = 'uniform')
    pred1 = knn.fit(X_train_scaler, y_train)
    pred = pred1.predict(X_test_scaler)
    accuracy = accuracy_score(y_test, pred)
    scores_uniform .append(accuracy)

    # KNN v·ªõi tr·ªçng s·ªë kho·∫£ng c√°ch
    knn = KNeighborsClassifier(n_neighbors = k, p = 2, weights = 'distance')
    pred2 = knn.fit(X_train_scaler, y_train)
    pred_ = pred2.predict(X_test_scaler)
    accuracy = accuracy_score(y_test, pred_)
    scores_distance.append(accuracy)


# B∆∞·ªõc 5: V·∫Ω bi·ªÉu ƒë·ªì so s√°nh
plt.figure(figsize=(10, 5))
plt.plot(n_neighbors, scores_uniform, label='weights=uniform')
plt.plot(n_neighbors, scores_distance, label='weights=distance')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Accuracy')
plt.title('Comparison of KNN with uniform vs distance weights')
plt.legend()
plt.grid(True)
plt.show()
```
